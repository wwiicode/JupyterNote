```python
import pandas as pd
X = pd.read_csv('agaricus-lepiota.csv', names = ['classes', 'cap-shape', 'cap-surface', 'cap-color', 'bruises?', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'], na_values = "?")
X.head()
```

```python
X[pd.isnull(X).any(axis=1)]
```

```python
X.dropna(axis = 0, how = 'any', inplace = True)
```

```python
y = X.classes
X.drop('classes', axis=1, inplace = True)
y = y.map({'e': 0, 'p': 1})
```

```python
X = pd.get_dummies(X, columns=['cap-shape', 'cap-surface', 'cap-color', 'bruises?', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'])
X.head()
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 7)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print "High-Dimensionality Score: ", round((score*100), 3)
```
